```python
%matplotlib Qt5
import matplotlib.pyplot as plt
from matplotlib import font_manager

import numpy as np

from Equation import Equation
# hype parameter

maxIter: int   = 500

fontsize = font_manager.FontProperties(size = 15)
tick_fontsize = font_manager.FontProperties(size = 12)
title_fontsize = font_manager.FontProperties(size = 14)
legned_fontsize = font_manager.FontProperties(size = 10)

row, col = 2, 2

fig = plt.figure(figsize=[10*col,3*row], dpi=256)

subgraph_item = 0 + ord('a') - 1
pic_idx = 0


item_pos = (-0.07, 1.15)
tight_layout_arg = dict(
    top=0.945,
    bottom=0.09,
    left=0.049,
    right=0.998,
    hspace=0.434,
    wspace=0.121
)
```

```python
# x2 - 2x - 5

tStart   : float = -100.0
interval : float = 1.0
tStep    : int   = 200
maxIter  : int   = 500

# src_label: str = r'$f(x) = e^{0.1x}$'; mathrm_src_label: str = r'$\mathrm{f(x) = e^{0.1x}}$'
src_label: str = r'$f(x) = x^2 - 2x - 5$'; mathrm_label: str = r'$\mathrm{f(x) = x^2 - 2x - 5}$'
# src_label: str = r'$f(x) = \sin x$'; mathrm_label: str = r'$\mathrm{f(x) = \sin x}$'
# src_label: str = r'$f(x) = \frac{1}{1 + e^{-x}}$'; mathrm_label: str = r'$\mathrm{f(x) = \frac{1}{1 + e^{-x}}}$'
# src_label: str = r'$f(x) = x^3$'; mathrm_label: str = r'$\mathrm{f(x) = x^3}$'

equation = Equation(tStart, tStep, interval, src_label)

trajs, derror, verror, traderror, traverror, mapederror, mapeverror = equation.init_loop(maxIter)

values = equation.values
label = equation.src

Intv = values - values[0]

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn generate trajectory derived from x0
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'Trajectory derived from $\\mathrm{{x_0}}$ = {tStart}', loc = 'center', fontproperties = title_fontsize)
plt.xlabel(r'$\mathrm{\Delta x}$', fontproperties = fontsize); # plt.ylabel('f(x)', fontproperties = fontsize)

plt.plot(Intv, trajs[:, 0], label = r'EdSr $\mathrm{f(\Delta x)}$', linewidth = 2)
plt.scatter(Intv, trajs[:, 0], s = 4)
plt.plot(Intv, label , label = r'$\mathrm{f(\Delta x) = (x_0 + \Delta x)^2 - 2(x_0 + \Delta x) - 5}$', linewidth = 1)
plt.scatter(Intv, label, s = 2)

plt.ticklabel_format(axis = 'y', style = 'sci', scilimits = (0, 2), useOffset = True, useMathText = True)

ax.yaxis.get_offset_text().set_fontsize(tick_fontsize.get_size())

plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'upper center', fontsize = legned_fontsize.get_size(), ncol = 1)


# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn Absolute value of error
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'AbsVE compared with {mathrm_label}', loc = 'center', fontproperties = title_fontsize)
plt.xlabel(r'$\mathrm{\Delta x}$', fontproperties = fontsize); # plt.ylabel('', fontproperties = fontsize)
plt.plot(Intv, derror, label = r'$\mathrm{f(\Delta x)}$', linewidth = 3)
plt.plot(Intv, verror, label = r'$\mathrm{\frac{df(\Delta x)}{dx}}$', linewidth = 1.5)
plt.axhline(y = 0., color = 'r', linestyle = '--', linewidth = 1, label = 'f(x) = 0.')
plt.yticks([-0.1, 0, 0.1])
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'lower right', fontsize = legned_fontsize.get_size(), ncol = 3)
```
