```python
# sin x

tStart   : float = -6.0
interval : float = 30.0
tStep    : int   = 100
maxIter  : int   = 50

# src_label: str = r'$f(x) = e^{0.1x}$'; mathrm_label: str = r'$\mathrm{f(x) = e^{0.1x}}$'
# src_label: str = r'$f(x) = x^2 - 2x - 5$'; mathrm_label: str = r'$\mathrm{f(x) = x^2 - 2x - 5}$'
src_label: str = r'$f(x) = \sin x$'; mathrm_label: str = r'$\mathrm{f(x) = \sin x}$'
# src_label: str = r'$f(x) = \frac{1}{1 + e^{-x}}$'; mathrm_label: str = r'$\mathrm{f(x) = \frac{1}{1 + e^{-x}}}$'
# src_label: str = r'$f(x) = x^3$'; mathrm_label: str = r'$\mathrm{f(x) = x^3}$'

equation = Equation(tStart, tStep, interval, src_label)

trajs, derror, verror, traderror, traverror, mapederror, mapeverror = equation.current_loop(maxIter)

values = equation.values
label = equation.src

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn generate trajectory derived from x(n-1)
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'$\\mathrm{{x_0}}$ = {tStart}, $\mathrm{{\Delta x}}$ = {interval}', loc = 'center', fontproperties = title_fontsize)
plt.xlabel(r'x', fontproperties = fontsize); # plt.ylabel('f(x)', fontproperties = fontsize)

plt.plot(values, trajs[:, 0], label = 'EdSr f(x)', linewidth = 2)
plt.scatter(values, trajs[:, 0], s = 4)
plt.plot(values, label , label = mathrm_label, linewidth = 1)
plt.scatter(values, label , s = 2)

plt.yticks([-1.0, -0.5, 0, 0.5, 1.0, 1.5])
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
plt.legend(loc = 'upper left', fontsize = legned_fontsize.get_size(), ncol = 2)
print('$\\mathit{x}$')

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn Absolute value of error
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'AbsVE compared with {mathrm_label}', loc = 'center', fontproperties = title_fontsize)
plt.xlabel(r'x', fontproperties = fontsize); # plt.ylabel('', fontproperties = fontsize)
plt.plot(values, derror, label = 'f(x)', linewidth = 2)
plt.plot(values, verror, label = r'$\mathrm{\frac{df(x)}{dx}}}$', linewidth = 1)
plt.yscale('log')
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'lower right', fontsize = legned_fontsize.get_size(), ncol = 2)
```
