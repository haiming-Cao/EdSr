```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager

import numpy as np

from idealSpring import IdealSpring
# hype parameter

maxIter  : int   = 500

fontsize = font_manager.FontProperties(size = 15)
tick_fontsize = font_manager.FontProperties(size = 10)
legned_fontsize = font_manager.FontProperties(size = 9.5)
title_fontsize = font_manager.FontProperties(size = 15)

row, col = 2, 2

fig = plt.figure(figsize=[20*col,3*row], dpi=256)

subgraph_item = 0 + ord('a') - 1

pic_idx = 0
item_pos = (-0.07, 1.15)

tight_layout_arg = dict(
    top=0.945,
    bottom=0.099,
    left=0.047,
    right=0.98,
    hspace=0.4,
    wspace=0.13
)
```

```python
tStart    = 0.0
interval  = 1.0
tStep     = 360

skip      = 3

spring = IdealSpring(tStart, tStep, interval)

trajs, derror, verror, traderror, traverror = spring.current_loop(maxIter)
times = spring.times

labelx = spring.labelx
labelv = spring.labelv

skip_trajs, skip_derror, skip_verror, skip_traderror, skip_traverror = trajs[::skip], derror[::skip], verror[::skip], traderror[::skip], traverror[::skip]
skip_times = times[::skip]
skip_labelx, skip_labelv = labelx[::skip], labelv[::skip]

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn generate trajectory derived from t(n-1)
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.xlabel(r't', fontproperties = fontsize); # plt.ylabel(r'$\mathrm{y_t/v_t}$', fontproperties = fontsize)
plt.title(f'$\\mathrm{{\Delta t}}$ = {interval}, $\mathrm{{t_0}}$ = {tStart}', loc = 'center', fontproperties = title_fontsize)
plt.plot(skip_times, skip_trajs[:, 0], label = r'EdSr y', linewidth = 2)
plt.scatter(skip_times, skip_trajs[:, 0], s = 4)
plt.plot(skip_times, skip_trajs[:, 1], label = r'EdSr v', linewidth = 2)
plt.scatter(skip_times, skip_trajs[:, 1], s = 4)
plt.plot(skip_times, skip_labelx, label = r'GT y', linewidth = 1)
plt.scatter(skip_times, skip_labelx , s = 2)
plt.plot(skip_times, skip_labelv, label = r'GT v', linewidth = 1)
plt.scatter(skip_times, skip_labelv , s = 2)

plt.yticks([-1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]); plt.ylim(top = 2.2)
plt.xticks(np.arange(tStart, tStart + interval * (tStep + 2), int(tStep // 9 * interval), dtype = np.int16))
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'upper left', fontsize = legned_fontsize.get_size(), ncol = 4)

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn Absolute Error derived from t(n-1)
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'AbsVE compared with GT', loc = 'center', fontproperties = title_fontsize)
plt.xlabel(r't', fontproperties = fontsize); # plt.ylabel(r'error', fontproperties = fontsize)
plt.plot(skip_times, skip_derror, label = r'EdSr y')
plt.plot(skip_times, skip_verror, label = r'EdSr v')
plt.plot(skip_times, skip_traderror, label = 'VV y')
plt.plot(skip_times, skip_traverror, label = 'VV v')

plt.yscale('log')
# plt.ylim(top = 1e-1)
# plt.yticks([1e-16, 1e-15, 1e-14])
# plt.yticks([1e-15, 1e-14, 1e-13]); plt.ylim(top = 2e-13)
plt.xticks(np.arange(tStart, tStart + interval * (tStep + 2), int(tStep // 9 * interval), dtype = np.int16))
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'center right', fontsize = legned_fontsize.get_size(), ncol = 2)
```
