```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager

import numpy as np

from idealSpring import IdealSpring
# hyper-parameter

maxIter  : int   = 50

fontsize = font_manager.FontProperties(size = 9)
tick_fontsize = font_manager.FontProperties(size = 8)
legned_fontsize = font_manager.FontProperties(size = 6)

row, col = 2, 2

fig = plt.figure(figsize=[20*col,6*row], dpi=400)

subgraph_item = 0 + ord('a') - 1

pic_idx = 0
item_pos = (-0.07, 1.15)

tight_layout_arg = dict(
    top=0.945,
    bottom=0.109,
    left=0.057,
    right=0.98,
    hspace=0.485,
    wspace=0.155
)
```

```python
tStart    = 0.0
interval  = 0.1
tStep     = 380

spring = IdealSpring(tStart, tStep, interval)

trajs, derror, verror, traderror, traverror = spring.init_loop(maxIter)
times = spring.times

timeIntv = spring.times - spring.times[0]

labelx = spring.labelx
labelv = spring.labelv

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn generate trajectory derived from t(0)
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.xlabel(r'$\mathrm{\Delta t}$', fontproperties = fontsize); # plt.ylabel(r'y/v($\mathrm{t_0 + \Delta t}$)', fontproperties = fontsize)
plt.title(f'trajectory derived from $\\mathrm{{t_0}}$ = {tStart}', loc = 'center', fontproperties = fontsize)
plt.plot(timeIntv, trajs[:, 0], label = r'EdSr y', linewidth = 2)
plt.scatter(timeIntv, trajs[:, 0], s = 4)
plt.plot(timeIntv, trajs[:, 1], label = r'EdSr v', linewidth = 2)
plt.scatter(timeIntv, trajs[:, 1], s = 4)

plt.plot(timeIntv, labelx, label = r'GT y', linewidth = 1)
plt.scatter(timeIntv, labelx , s = 2)
plt.plot(timeIntv, labelv, label = r'GT v', linewidth = 1)
plt.scatter(timeIntv, labelv , s = 2)

plt.yticks(np.arange(-1.5, 2.0 + 0.5, 0.5))
plt.xticks(np.arange(timeIntv[0], timeIntv[-1] + int(tStep // 9 * interval), int(tStep // 9 * interval), dtype = np.int16))
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'upper left', fontsize = legned_fontsize.get_size(), ncol = 4)

# ! Picture idx
pic_idx += 1
subgraph_item += 1

ax = plt.subplot(row, col, pic_idx)
# attn Absolute value of Error derived from t(0)
plt.text(*item_pos, chr(subgraph_item), transform = ax.transAxes, fontsize = fontsize.get_size(), fontweight = 'bold', va = 'top', ha = 'left')
plt.title(f'AbsVE compared with GT', loc = 'center', fontproperties = fontsize)
plt.xlabel(r'$\mathrm{\Delta t}$', fontproperties = fontsize); # plt.ylabel(r'error', fontproperties = fontsize)
plt.plot(timeIntv, derror, label = r'EdSr y')
plt.plot(timeIntv, verror, label = r'EdSr v')
plt.plot(timeIntv, traderror, label = r'VV y')
plt.plot(timeIntv, traverror, label = r'VV v')
plt.yscale('log')
plt.xticks(np.arange(timeIntv[0], timeIntv[-1] + int(tStep // 9 * interval), int(tStep // 9 * interval), dtype = np.int16))
plt.tick_params(axis = 'both', labelsize = tick_fontsize.get_size())
if tight_layout_arg is not None:
    plt.subplots_adjust(**tight_layout_arg)
plt.legend(loc = 'lower right', fontsize = legned_fontsize.get_size())
```
